# Missing Data & Data Normalization

## Checking Existing Null Value & Missing Value

```python
display(csv_data.isnull().values.any())
display(csv_data.isnull().sum())
```

## Handling  Missing Value

```python
csv_data_missing_value = pd.read_csv("https://storage.googleapis.com/dqlab-dataset/shopping_data_missingvalue.csv")
print("Checking Existing Missing Value or Null Value = ?")
display(csv_data_missing_value.isnull().sum())
```

### Imputation : Mean()

```python
csv_data = csv_data_missing_value.copy()

print("Dataset that still have Missing Value = ?")
display(csv_data.isnull().sum())
display(csv_data.select_dtypes("number").mean())

missing_value_columns = csv_data.columns[csv_data.isnull().any()]
csv_data[missing_value_columns] = csv_data[missing_value_columns].fillna(csv_data[missing_value_columns].mean())

print("Dataset that still have Missing Value = ?")
display(csv_data.isnull().sum())
display(csv_data.select_dtypes("number").mean())
```

### imputation : Median()

```python
csv_data = csv_data_missing_value.copy()

print("Dataset that still have Missing Value = ?")
display(csv_data.isnull().sum())
display(csv_data.select_dtypes("number").mean())

missing_value_columns = csv_data.columns[csv_data.isnull().any()]
csv_data[missing_value_columns] = csv_data[missing_value_columns].fillna(csv_data[missing_value_columns].median())

print("Dataset that still have Missing Value = ?")
display(csv_data.isnull().sum())
display(csv_data.select_dtypes("number").mean())
```

## Data Normalization

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
csv_data = pd.read_csv("https://storage.googleapis.com/dqlab-dataset/shopping_data.csv")
display(csv_data.head())
```

### Data Normalization Using Scikit Learn

```python
csv_data_normalization = csv_data.copy()
min_max_scaler = MinMaxScaler()

scaller_columns = csv_data_normalization.columns[2:]

csv_data_normalization[scaller_columns] = min_max_scaler.fit_transform(csv_data_normalization[scaller_columns])

display(csv_data_normalization.head())
```
